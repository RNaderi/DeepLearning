```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import plot_model

import os
os.environ["PATH"] += os.pathsep + 'D:/Program Files (x86)/Graphviz2.38/bin/'
```

```python
X,Y=datasets.make_classification(n_samples=200,n_classes=2,n_clusters_per_class=1,n_features=2,
                                 n_informative=2
                                 ,n_redundant=0
                                ,n_repeated=0) #(200,2)
x_train,x_test,y_train,y_test= train_test_split(X, Y , test_size=0.3)
c=y_train
plt.figure(0)
plt.title('DataSet')
plt.xlabel('x_train0')
plt.ylabel('x_train1')
plt.scatter(x_train[:,0],x_train[:,1],c=c,marker='o')   
```

```python
y_train=y_train.reshape(-1,1)
n_input=x_train.shape[1] #feature_number
n_outputs=1
m=x_train.shape[0]   #X(200,2)
learning_rate=0.01
n_epoch=500
```

```python
#create model 
model=Sequential()
#input_dim=2 is equivalent to input_shape=(2,)
output_layer=Dense(units=1,input_dim=2, activation='sigmoid' , name="output_layer" )
#config = output_layer.get_config()

model.add(output_layer)  #add model
model.compile(loss='binary_crossentropy' ,optimizer='adam' ,metrics=['accuracy'])
model.summary()
plot_model(model=model,to_file='logistic_kerasModel.png',show_shapes=True)
```

```python
#train the model
model_history=model.fit(x_train,y_train,epochs=n_epoch)
```

```python
#plot loss and accuracy
loss_values=model_history.history['loss']
plt.figure()
plt.title('loss_values')
plt.plot(loss_values)


accuracy_values=model_history.history['acc']
plt.figure()
plt.title('accuracy_values')
plt.plot(accuracy_values)
```

```python
#plot line
weights=output_layer.get_weights()   # returns the weights of the layer as a list of Numpy arrays.
w1,w2=weights[0]
bias =weights[1]
bias,w1,w2


#plot
plt.figure(1)
plt.title('DataSet')
plt.xlabel('x_train0')
plt.ylabel('x_train1')
plt.scatter(x_train[:,0],x_train[:,1],c=c,marker='o')   

x_1 = np.linspace(np.min(x_train),np.max(x_train),x_train.shape[0])
#x_1 = np.linspace(np.min(x_train),np.max(x_train),x_train.shape[0])
x_2 = -(x_train[:,0]*w1[0]+bias)/w2[0]
plt.plot(x_1,x_2);   
```

```python
#evaluation model on a test set
[loss,accuracy]=model.evaluate(x_test,y_test)  #Returns the loss value & metrics values for the model in test mode.
print("loss: ",loss, "\n accuracy:" ,accuracy) 
```

```python
#predict class (a number between 0 and 1 : sigmoid)
y_predict=model.predict(x_test)
np.round(y_predict)
#y_predict1=model.predict_classes(x_test)
#np.equal(np.round(y_predict),y_predict1)
```
